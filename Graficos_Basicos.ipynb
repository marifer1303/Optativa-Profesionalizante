```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#grafico de lineas
plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.show()
```

```python
#Grafico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]
plt.figure(figsize=(10, 4))
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
#Grafico de dispersion
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 4))
plt.scatter(x, y)
plt.title('Gráfico de Dispersión')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#histograma
datos = np.random.randn(1000)

plt.figure(figsize=(10, 4))
plt.hist(datos, bins=30)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```
