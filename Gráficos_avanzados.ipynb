# Heatmap

# USOS:
-  Visualizar la **densidad o intensidad** de los datos en una matriz, permitiendo identificar **patrones, correlaciones y variaciones** en un conjunto de datos.
- V**isualización de Correlaciones:** En análisis estadísticos, los heatmaps se utilizan para mostrar la correlación entre variables. Cada celda del mapa representa el coeficiente de correlación entre dos variables, y el color indica la fuerza y dirección de la correlación.
- **Análisis de Clusters:** En biología y bioinformática, los heatmaps son comunes para representar datos de expresión génica, donde se pueden identificar clusters de genes con patrones similares.
- **Visualización de Matrices:** Se utilizan para visualizar matrices grandes en diversas aplicaciones científicas e ingenieriles.
- **Análisis Geoespacial:** En mapas geográficos, los heatmaps pueden mostrar la densidad de eventos o características en diferentes áreas geográficas.
- **Representación de Datos Temporales:** Pueden mostrar cómo cambian ciertos valores a lo largo del tiempo en diferentes categorías o ubicaciones.

# INTERPRETACION:

- **Colores**: Los colores representan la magnitud o intensidad de los valores. Por ejemplo, en un heatmap de correlación:
 - Colores cálidos (como rojo) pueden indicar una fuerte correlación positiva.
 - Colores fríos (como azul) pueden indicar una fuerte correlación negativa.
 - Colores neutros (como blanco o gris) pueden indicar poca o ninguna correlación.
- **Patrones y Clusters:** Al observar el mapa, se pueden identificar patrones o agrupaciones donde ciertos valores son similares. Esto puede indicar relaciones interesantes entre las variables.
- **Ejes y Etiquetas:** Los ejes suelen estar etiquetados con las variables que se están comparando. Es importante prestar atención a estas etiquetas para entender qué relaciones se están visualizando.

plt.figure(figsize=(10, 6)): Crea una nueva figura de Matplotlib. figsize=(10, 6) establece el tamaño de la figura en pulgadas (ancho: 10, alto: 6).

sns.violinplot(x="day", y="total_bill", data=tips): sns.violinplot() es la función de Seaborn para crear gráficos de violín.
 - x="day": Especifica la variable categórica para el eje x (los días).
 - y="total_bill": Especifica la variable numérica para el eje y (el total de la cuenta).

# USOS:
- Muestra la distribución de datos cuantitativos a través de diferentes categorías.
  - **Comparación de Distribuciones:** Se utiliza para comparar la distribución de una variable numérica entre diferentes grupos o categorías. Es útil cuando se quiere entender cómo varían los datos dentro de cada categoría.
  - **Visualización de la Densidad:** Muestra la densidad de los datos a lo largo del rango de valores, lo que permite identificar no solo la mediana y los cuartiles (como en un box plot), sino también la forma completa de la distribución.
  - **Identificación de Modas:** Permite visualizar múltiples modas en los datos, es decir, picos en la densidad que indican valores más frecuentes.

# INTERPRETACIÓN:

- **Forma del Violín:** La anchura del violín en cualquier punto indica la densidad de los datos en ese valor. Partes más anchas indican mayor densidad (más datos) y partes más estrechas indican menor densidad.
- **Mediana y Cuartiles:** Generalmente, el gráfico incluye una línea central que representa la mediana y líneas adicionales que representan los cuartiles, similar a un box plot.
- **Simetría y Asimetría:** puede indicar si los datos están distribuidos uniformemente o si están sesgados hacia un lado.
- **Outliers:** Aunque no siempre se muestran explícitamente en un violin plot, los outliers pueden inferirse si hay extensiones largas y estrechas en los extremos del violín.

Este código:

- import matplotlib.cm as cm: Importa el módulo de mapas de colores (colormap) de Matplotlib y lo renombra como cm.
- plt.figure(figsize=(10, 6)): Crea una nueva figura de Matplotlib con un tamaño de 10x6 pulgadas.
- n_categories = len(tips['day'].
 - unique()): Cuenta el número único de categorías en la columna 'day' del DataFrame 'tips'.
- colors = cm.rainbow(np.linspace(0, 1, n_categories)): Crea un array de colores usando el mapa de colores 'rainbow'.
 - np.linspace(0, 1, n_categories) genera números espaciados uniformemente entre 0 y 1.
- colors_list = [tuple(c) for c in colors]: Convierte el array de colores en una lista de tuplas RGB.
- sns.violinplot(x="day", y="total_bill", hue="day", data=tips, palette=colors_list, legend=False): Crea el gráfico de violín usando Seaborn.
 - x="day": Especifica la variable categórica para el eje x.
 - y="total_bill": Especifica la variable numérica para el eje y.
 - hue="day": Usa la misma variable que el eje x para determinar el color.
 - data=tips: Indica el DataFrame que contiene los datos.
 - palette=colors_list: Usa la lista de colores personalizada.
 - legend=False: Evita que se muestre la leyenda.
- plt.title('Violin plot: Total de la cuenta por día'): Añade un título al gráfico.

```python
import matplotlib.cm as cm

plt.figure(figsize=(10, 6))
n_categories = len(tips['day'].unique())
colors = cm.rainbow(np.linspace(0, 1, n_categories))
colors_list = [tuple(c) for c in colors]  # Convertir a lista de tuplas

sns.violinplot(x="day", y="total_bill", hue="day", data=tips,
               palette=colors_list, legend=False)
plt.title('Violin plot: Total de la cuenta por día')
plt.show()
```

- x="day": Especifica la variable categórica para el eje x (los días).
- y="total_bill": Especifica la variable numérica para el eje y (el total de la cuenta).
- hue="day": Usa la misma variable que el eje x para determinar el color de cada violín.
- data=tips: Indica el DataFrame de pandas que contiene los datos.

# Pair plot

sns.pairplot(tips, hue="time"): sns.pairplot() es una función de Seaborn que crea una matriz de gráficos de dispersión para todas las combinaciones posibles de variables numéricas en el DataFrame.
- tips es el DataFrame que contiene los datos.
- hue="time" especifica que los puntos en los gráficos se colorearán según la variable "time" (que podría ser, por ejemplo, almuerzo o cena).

plt.suptitle('Pair plot de variables en el conjunto de datos tips', y=1.02): Añade un título principal (supertítulo) al conjunto completo de gráficos.
- 'Pair plot de variables en el conjunto de datos tips' es el texto del título.
- y=1.02 ajusta la posición vertical del título, colocándolo ligeramente por encima de la figura.

Este código creará una matriz de gráficos de dispersión donde:
* Cada variable numérica en el DataFrame tips se graficará contra todas las demás variables numéricas.
* La diagonal de la matriz mostrará histogramas o gráficos de densidad para cada variable.
* Los puntos en cada gráfico de dispersión se colorearán según la variable "time".
* Habrá un título general en la parte superior del conjunto de gráficos.


**USOS:**
- Explorar relaciones entre múltiples variables en un conjunto de datos
- Para identificar patrones y correlaciones potenciales.
- Visualización de distribuciones de cada variable individual en la diagonal del gráfico.
- Proporciona una visión general de cómo se relacionan múltiples variables entre sí.

**INTERPRETACIÓN:**

- **Matriz de gráficos de dispersión:** Cada celda fuera de la diagonal representa un gráfico de dispersión entre dos variables.
 - El eje x de cada gráfico es la variable de la columna correspondiente.
 - El eje y es la variable de la fila correspondiente.
- **Diagonal:** Muestra la distribución de cada variable individual, generalmente como un histograma o un gráfico de densidad.
-**Correlaciones:** Patrones lineales en los gráficos de dispersión sugieren correlaciones.
  - Líneas ascendentes indican correlaciones positivas.
  - Líneas descendentes indican correlaciones negativas.
- **Agrupaciones:** Si se usa el parámetro 'hue', los puntos se colorearán según una variable categórica, permitiendo identificar patrones o diferencias entre grupos.
- **Outliers:** Puntos aislados o alejados del patrón general pueden indicar valores atípicos.
- **Formas no lineales:** Curvas o patrones no lineales en los gráficos de dispersión sugieren relaciones más complejas entre variables.
- **Densidad de puntos:** Áreas con mayor concentración de puntos indican valores más comunes o típicos.
- **Rangos y escalas:** Observa los rangos de valores en los ejes para entender la escala y distribución de cada variable.


```python
plt.figure(figsize=(12, 10))
sns.pairplot(tips, hue="time")
plt.suptitle('Pair plot de variables en el conjunto de datos tips', y=1.02)
plt.show()
```

# Box plot

**USOS:**
- visualizar la distribución de un conjunto de datos numéricos.
- Mostrar la distribución de los datos.
- Identificar valores atípicos (outliers).
- Comparar distribuciones entre diferentes grupos o categorías.
- Visualizar la simetría o asimetría de los datos.
- Proporcionar una visión rápida de las medidas de tendencia central y dispersión.

**INTERPRETACIÓN:**
- La **caja:** Representa el rango intercuartílico (IQR), que contiene el 50% central de los datos.
 - El borde inferior de la caja es el primer cuartil (Q1, 25% de los datos).
 - El borde superior de la caja es el tercer cuartil (Q3, 75% de los datos).
- La **línea dentro de la caja:** Representa la mediana (segundo cuartil, Q2). Su posición indica si los datos son simétricos (en el centro) o asimétricos.
- Los **bigotes:** Se extienden desde la caja hasta los valores mínimo y máximo dentro de 1.5 veces el IQR. Muestran la dispersión de los datos fuera del rango intercuartílico.
- **Puntos individuales** más allá de los bigotes: Representan valores atípicos (outliers). Indican datos que están inusualmente lejos del resto de la distribución.
- **Longitud** de la caja y bigotes: Una caja larga indica mayor dispersión de los datos centrales. Bigotes largos sugieren una mayor dispersión en los extremos de la distribución.
- **Comparación** entre grupos: Cuando se muestran varios boxplots lado a lado, se pueden comparar fácilmente las distribuciones de diferentes grupos o categorías.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
plt.title('Box plot: Total de la cuenta por día y si es fumador')
plt.show()
```

# Regresión lineal

**USOS:**
- Visualizar la relación entre dos variables continuas.
- Mostrar la tendencia general de los datos.
- Identificar patrones lineales o no lineales en los datos.
- Evaluar visualmente la fuerza de la relación entre variables.
- Detectar valores atípicos o outliers.

**INTERPRETACIÓN:**
- Nube de puntos: Cada punto representa una observación individual. La posición de los puntos muestra la relación entre las variables x e y.
- Línea de regresión: La línea recta representa el mejor ajuste lineal a los datos. Una línea ascendente indica una relación positiva, mientras que una descendente indica una relación negativa. Cuanto más inclinada sea la línea, más fuerte es la relación entre las variables.
- Área sombreada: Representa el intervalo de confianza alrededor de la línea de regresión. Un intervalo de confianza más estrecho indica una estimación más precisa.
- Dispersión de puntos alrededor de la línea: Si los puntos están cerca de la línea, sugiere una fuerte relación lineal. Si los puntos están muy dispersos, indica una relación más débil o posiblemente no lineal.
- Outliers: Puntos que están muy alejados de la tendencia general pueden ser outliers.
- Forma de la relación: Si los puntos parecen seguir un patrón curvo en lugar de lineal, podría indicar una relación no lineal entre las variables.
- Densidad de puntos: Áreas con mayor concentración de puntos indican valores más comunes o típicos

```python

plt.figure(figsize=(10, 6))
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title('Regresión lineal: Propina vs Total de la cuenta')
plt.show()
```
